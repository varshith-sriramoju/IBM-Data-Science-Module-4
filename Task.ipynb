```python
import yfinance as yf

# Create Tesla ticker object
tesla = yf.Ticker("TSLA")

# Get historical stock data for maximum period
tesla_data = tesla.history(period="max")

# Reset index and display first 5 rows
tesla_data.reset_index(inplace=True)
tesla_data.head()
```

```python
import pandas as pd

# Sample Tesla revenue data (manual entry)
data = {
    'Year': [2021, 2022, 2023, 2024, 2025],
    'Revenue ($M)': [53823, 81462, 107213, 120000, 130000]
}
tesla_revenue = pd.DataFrame(data)

# Display last 5 rows
tesla_revenue.tail()
```

```python
# Create GameStop ticker object
gme = yf.Ticker("GME")

# Get historical stock data for maximum period
gme_data = gme.history(period="max")

# Reset index and display first 5 rows
gme_data.reset_index(inplace=True)
gme_data.head()
```

```python
# Sample GameStop revenue data
data = {
    'Year': [2021, 2022, 2023, 2024, 2025],
    'Revenue ($M)': [6011, 5927, 5900, 6000, 6100]
}
gme_revenue = pd.DataFrame(data)

# Display last 5 rows
gme_revenue.tail()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(tesla_data['Date'], tesla_data['Close'], label='Tesla Close Price')
plt.title('Tesla Stock Closing Price Over Time')
plt.xlabel('Date')
plt.ylabel('Close Price ($)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(gme_data['Date'], gme_data['Close'], label='GameStop Close Price', color='red')
plt.title('GameStop Stock Closing Price Over Time')
plt.xlabel('Date')
plt.ylabel('Close Price ($)')
plt.legend()
plt.show()
```
